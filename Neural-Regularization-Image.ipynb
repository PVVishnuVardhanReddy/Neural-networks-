```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten,Dropout
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns

(X_train, y_train), (X_test, y_test) = mnist.load_data()

X_train = X_train.astype('float32') / 255.0
X_test = X_test.astype('float32') / 255.0

X_train_flattened = X_train.reshape((X_train.shape[0], -1))
X_test_flattened = X_test.reshape((X_test.shape[0], -1))

y_train_encoded = to_categorical(y_train)
y_test_encoded = to_categorical(y_test)

def build_model(hidden_layers, activation_fn, regularization=None):
    model = Sequential()
    for units in hidden_layers:
        if regularization == 'l1':
            model.add(Dense(units=units, activation=activation_fn, kernel_regularizer=tf.keras.regularizers.l1(0.001)))
        elif regularization == 'l2':
            model.add(Dense(units=units, activation=activation_fn, kernel_regularizer=tf.keras.regularizers.l2(0.001)))
        elif regularization=="dropout":
            model.add(Dropout(0.30))
        else:
            model.add(Dense(units=units, activation=activation_fn))
    model.add(Dense(units=10, activation='softmax'))
    return model

def train_and_evaluate_model(model, X_train, y_train, X_test, y_test, hidden_layers, activation_fn, regularization):
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    model.fit(X_train, y_train, epochs=10, batch_size=32, verbose=0)
    y_pred = np.argmax(model.predict(X_test), axis=-1)
    test_accuracy = np.mean(y_pred == np.argmax(y_test, axis=-1))
    print(f"Model Configuration: Hidden Layers: {hidden_layers}, Activation Function: {activation_fn}, Regularization: {regularization}")
    print("Classification Report:")
    print(classification_report(np.argmax(y_test, axis=-1), y_pred))
    print("Confusion Matrix:")
    cm = confusion_matrix(np.argmax(y_test, axis=-1), y_pred)
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=range(10), yticklabels=range(10))
    plt.xlabel("Predicted labels")
    plt.ylabel("True labels")
    plt.title("Confusion Matrix")
    plt.show()
    return test_accuracy

hidden_layers_configs = [[32],[64]]

activation_functions = ['relu', 'sigmoid', 'tanh']
regularization_methods = ['l1', 'l2',"dropout"]

accuracies = []
config_labels = []

for hidden_layers in hidden_layers_configs:
    for activation_fn in activation_functions:
        for reg_method in regularization_methods:
            model = build_model(hidden_layers, activation_fn, regularization=reg_method)
            accuracy = train_and_evaluate_model(model, X_train_flattened, y_train_encoded, X_test_flattened, y_test_encoded, hidden_layers, activation_fn, reg_method)
            accuracies.append(accuracy)
            config_labels.append(f"Hidden Layers: {hidden_layers}, Activation Function: {activation_fn}, Regularization: {reg_method}")

plt.figure(figsize=(12, 8))
plt.barh(config_labels, accuracies, color='skyblue')
plt.xlabel('Accuracy')
plt.ylabel('Model Configuration')
plt.title('Accuracy of Models with Different Configurations, Activation Functions, and Regularization on MNIST Dataset')
plt.xlim(0, 1)
plt.gca().invert_yaxis()
plt.show()
```
